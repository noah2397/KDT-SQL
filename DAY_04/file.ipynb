```python
#readlines() 함수
#• 각 라인이 저장된 리스트를 반환
infile = open("phones.txt", "r")
lines = infile.readlines()
print(lines)
for line in lines:
    print(line)
infile.close()
```

```python
# 파일의 끝(End of File: EOF) 구분
#• EOF를 만나면, empty string(‘’)을 리턴함
infile = open("phones.txt", "r", encoding='utf-8')
line = infile.readline()

while line != "": # EOF가 아닐 때까지 반복
    print(line, end='')
    line = infile.readline()
infile.close()
```

```python
#str.rstrip() 함수
#• 문자열의 오른쪽 공백 문자를 제거하는 메소드
#• 파일에서 줄 마지막에 붙은 ‘\n'을 제거할 때 유용함
infile = open("phones.txt", "r")
for line in infile:
    line = line.rstrip()
    print(line)
infile.close()
```

```python
# 파일에서 문자 단위로 읽기
#• read(1): 한 문자씩 읽음
infile = open("input.txt", "r")
ch = " "
while ch != "" :
    ch = infile.read(1)
    print(ch, end='')
infile.close()
```

```python
# 바이너리 파일(이미지파일) 복사


filename1 = input("원본 파일 이름을 입력하시오: ")
filename2 = input("복사 파일 이름을 입력하시오: ")
infile = open(filename1, "rb")
outfile = open(filename2, "wb")
# 입력 파일에서 1024 바이트씩 읽어서 출력 파일에 쓴다.
# 파일의 마지막 부분에서는 읽어 들인 바이트 수만큼 파일에 저장
while True:
    copy_buffer = infile.read(1024)
    print(len(copy_buffer)) # 실제 읽어온 바이트 수 출력
    if not copy_buffer: # 파일의 끝인 경우, empty byte를 리턴
        break

    outfile.write(copy_buffer)
infile.close()
outfile.close()
print(filename1+"를 " + filename2 +"로 복사하였습니다. ")
```

디렉토리 관련 함수
• os.getcwd(): 현재 작업 디렉토리의 경로 얻기
– cwd: current working directory

• os.chdir(경로명): 작업 디렉토리 변경

• os.listdir(): 작업 디렉토리 안에 있는 파일들과 디렉토리 이름 반환

§ 파일 관련
• os.path.isfile(파일이름): 파일 존재 여부 확인

• os.path.isdir(디렉토리이름): 디렉토리 존재 여부 확인

• str.endswith(“확장자”): 파일 확장자 검사

```python
import os
dir = os.getcwd()
print(dir)
subdir = "../DAY_04"
os.chdir(subdir)
print(os.getcwd())
for filename in os.listdir():
    print(filename, end='')
    if os.path.isfile(filename):
        print(": 파일")
    elif os.path.isdir(filename):
        print(": 디렉토리")
```

```python
pickle 모듈 사용
• dump()와 load() 메소드를 사용하여 바이트 형태로 객체를 쓰고 읽을 수 있음

§ pickle.dump(obj, file)
• obj 형태를 파일 객체(file)에 저장함
• 객체를 바이트 스트림으로 변환해서 파일로 저장: 직렬화

§ pickle.load(file)
• 바이트 스트림이 객체 구조로 변환: 역직렬화
• 한 줄씩 파일을 읽어오고 더 이상 로드할 데이터가 없으면 EOFError 발생

§ 사용 형식
pickle.dump(데이터, 파일) # 객체를 파일로 저장
객체 or 자료형 변수 = pickle.load(파일) # 파일에서 객체를 읽어옴
```

```python
import pickle
class Person(object):
    def __init__ (self, id, name):
        self.id=id
        self.name=name
    def __repr__(self):
        return f"({self.id}{self.name})"
p1=Person(1,"Kim")
p2=Person(2, "Park")
p3=Person(3,"Lee")

person_list=[p1,p2,p3]

# 객체 리스트를 dump를 사용하여 파일로 저장
fname="person.pickle" # pickle로 객체를 저장했음을 보여주는 보여주기식 파일 확장자이다
fout=open(fname, "wb")
pickle.dump(person_list, fout)
fout.close()

# 파일의 내용을 객체 리스트로 읽어옴
fin=open(fname, "rb")
plist=pickle.load(fin)
print(plist)
fin.close()
```

```python
# 다중 예외 처리 구조
try:
    a = [1,2]
    print(a[3])
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except IndexError:
    print("인덱싱 할 수 없습니다.")
```

```python
# finally 블록 실행
#• 예외 발생 여부와 관계없이 항상 수행
#• 반드시 수행되어야 되는 문장을 추가
def calc(values):
    sum = None
    try:
        sum = values[0] + values[1] + values[2]
    except IndexError as err:
        print("인덱스 에러: " , err)
    except Exception as err:
        print(err)
    else:
        print("에러 없음:", values)
    finally:
        print(f"sum={sum}")

calc([1, 2, 3])
calc([1, 2])
```
